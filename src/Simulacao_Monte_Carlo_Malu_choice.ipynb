# Coluna "Number of Form Submissions"


```python
df['Became a Lead Date'] = pd.to_datetime(df['Became a Lead Date'])
dados_antes_whats = df[df['Became a Lead Date'] < '2024-04-18']
dados_depois_whats = df[df['Became a Lead Date'] >= '2024-04-18']
```

```python
max_forms_submetidos_antes = dados_antes_whats['Number of Form Submissions'].max()
min_forms_submetidos_antes = dados_antes_whats['Number of Form Submissions'].min()

max_forms_submetidos_depois = dados_depois_whats['Number of Form Submissions'].max()
min_forms_submetidos_depois = dados_depois_whats['Number of Form Submissions'].min()

print("Maior valor de formulários submetidos antes:", max_forms_submetidos_antes)
print("Menor valor de formulários submetidos antes:", min_forms_submetidos_antes)
print("Maior valor de formulários submetidos depois:", max_forms_submetidos_depois)
print("Menor valor de formulários submetidos depois:", min_forms_submetidos_depois)
```

```python
def remove_outliers(data):
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return data[(data >= lower_bound) & (data <= upper_bound)]

forms_submetidos_sem_outliers_antes = remove_outliers(dados_antes_whats['Number of Form Submissions'])
forms_submetidos_sem_outliers_depois = remove_outliers(dados_depois_whats['Number of Form Submissions'])

todos_dados_sem_outliers_forms_submetidos = pd.concat([forms_submetidos_sem_outliers_antes, forms_submetidos_sem_outliers_depois])
```

```python
histograma, bin_edges = np.histogram(todos_dados_sem_outliers_forms_submetidos, bins=30)

total = len(todos_dados_sem_outliers_forms_submetidos)
porcentagens = [count / total * 100 for count in histograma]

fig = px.bar(x=bin_edges[:-1], y=porcentagens,
             title='Distribuição do Número de Formulários Submetidos (Sem Outliers)',
             labels={'x': 'Número de Formulários Submetidos', 'y': 'Porcentagem'},
             opacity=0.7, color_discrete_sequence=['green'])

fig.update_xaxes(type='linear', tickmode='linear', tick0=0, dtick=1)

fig.show()
```

```python
def simulacao(valor):
    resultado = valor
    return resultado

array = np.array([1, 2, 3, 4])

probabilidades = np.array([0.04, 0.56, 0.3, 0.1])

num_simulacoes = 100

resultados_simulacao = []

for _ in range(num_simulacoes):
    amostra = np.random.choice(array, size=3, replace=True, p=probabilidades)

    resultado_simulacao = simulacao(amostra)

    resultados_simulacao.append(resultado_simulacao)

fig = px.histogram(x=resultados_simulacao,
                   title='Histograma dos Resultados das Simulações de Monte Carlo',
                   labels={'x': 'Resultado da Simulação', 'y': 'Frequência'},
                   opacity=0.7, color_discrete_sequence=['pink'])
fig.update_layout(bargap=0.05)
fig.show()
```
